```python
matches_played=pd.concat([match_df['team2'],match_df['team2']])
matches_played=matches_played.value_counts().reset_index()
matches_played.columns=['Team','Total Matches']
matches_played['wins']=match_df['winner'].value_counts().reset_index()['winner']

matches_played.set_index('Team',inplace=True)
```

```python
matches_played
```

```python
win_percentage=(matches_played['wins']/matches_played['Total Matches'])*100
```

```python
win_percentage
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title("Total Matches played Vs Percentage Match Won")
sns.barplot(matches_played['Total Matches'].index,win_percentage);
```

```python
plt.subplots(figsize=(20,10))
plt.title("WIN Percentage by Teams")
plt.pie(win_percentage,labels=win_percentage.index,autopct='%1.1f%%',shadow = True, startangle = 90);
plt.tight_layout()
```

# The team with the most number of wins per season

```python
ipl_teams_season=match_df.groupby('season')['winner'].value_counts()
```

```python
ipl_teams_season
```

```python
year = 2008
win_per_season = pd.DataFrame(columns=['year', 'team', 'wins'])
for items in ipl_teams_season.iteritems():    
    if items[0][0]==year:
        print(items)
        win_series = pd.DataFrame({
            'year': [items[0][0]],
            'team': [items[0][1]],
            'wins': [items[1]]
        })
        win_per_season = win_per_season.append(win_series)
        year += 1   
```

```python
win_per_season
```

```python
plt.subplots(figsize=(15,10))
plt.title("Per Season Top Teams")
sns.barplot(x='wins',y='team',hue='year',data=win_per_season)
```

                    Conclusion:  Mumbai indians is the most successful  teams

##  Best Bowlers in the history of ipl

```python
deliveries_df.head()
```

```python
### Most spells by an bowler in ipl 
most_spells=deliveries_df['bowler'].value_counts()[:20]
most_spells
sns.barplot(most_spells,most_spells.index);
```

```python
bowling_wickets = deliveries_df[deliveries_df['dismissal_kind']!='run out']
bowling_total = bowling_wickets.groupby('bowler').apply(lambda x : x['dismissal_kind'].dropna()).reset_index(name='Wickets')
bowling_wicket_count = bowling_total.groupby('bowler').count().reset_index()
bowling_top = bowling_wicket_count.sort_values(by='Wickets',ascending=False)
top_bowlers = bowling_top.loc[:,['bowler','Wickets']][0:10]
top_bowlers
```

```python
plt.figure(figsize=(12,6))
plt.scatter(top_bowlers['bowler'],top_bowlers['Wickets'],color='r');
plt.plot(top_bowlers['bowler'],top_bowlers['Wickets'],color='g');
plt.xticks(rotation=60)
plt.xlabel('Top 10 Bowlers')
plt.ylabel('Wickets Taken')
plt.title('Top 10 Bowlers in ipl history');
```

 Here SL Malinga is the most wicket taker and 7 indian bowlers in this list

```python
batsmen = deliveries_df.groupby('batsman').apply(lambda x : np.sum(x['batsman_runs'])).reset_index(name='Runs')
top_batsmen=batsmen.sort_values(by='Runs',ascending=False)[:10]
top_batsmen
```

```python
plt.figure(figsize=(15,8))
plt.title('Top 10 Batsmen')
sns.barplot(top_batsmen['batsman'],top_batsmen['Runs']);
```

Virat Kohli is the highest run getter, 7 indian Batsmen in Top 10 list
